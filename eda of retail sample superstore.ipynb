```python
x=df['State']
y=df['Sales']
plt.figure(figsize=(25,15))
plt.scatter(x,y,color="k")
plt.xlabel('State')
plt.ylabel('Sales')
plt.title(' scatter No of sales in a specific state')
plt.show()
```
